```python
#scatter plot

# Sample data from the image
dates = ['Jul-05', 'Aug-05', 'Sep-05', 'Oct-05', 'Nov-05','Dec-05']
close_prices = [12.46, 13.42, 13.3, 12.99, 13.41]

# Plotting the scatter plot
plt.scatter(dates, close_prices, color='blue')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Sample Stock Prices')
plt.xticks(rotation=45)
plt.show()
```

```python
#Bar chart

# Sample data from the image
dates = ['Jul-05', 'Aug-05', 'Sep-05', 'Oct-05', 'Nov-05','Dec-05']
close_prices = [12.46, 13.42, 13.3, 12.99, 13.41,13.71]

# Creating the bar chart
plt.bar(dates, close_prices, color='pink')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Sample Stock Close Prices')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjusts plot to ensure everything fits without overlap
plt.show()
```

```python
#Pie chart
# Sample data from the image
dates = ['Jul-05', 'Aug-05', 'Sep-05', 'Oct-05', 'Nov-05','Dec-05']
close_prices = [12.46, 13.42, 13.3, 12.99, 13.41,13.71]

# Creating the pie chart
plt.figure(figsize=(8, 8))
plt.pie(close_prices, labels=dates, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Close Prices (Sample Data)')
plt.show()
```

```python
#Heatmap
# Sample data extracted from the image
data = {
    'Date': ['Jul-05', 'Aug-05', 'Sep-05', 'Oct-05', 'Nov-05'],
    'Open': [13, 12.58, 13.48, 13.2, 13.35],
    'High': [14, 14.88, 14.87, 14.47, 13.88],
    'Low': [11.25, 12.55, 12.27, 12.4, 12.88],
    'Close': [12.46, 13.42, 13.3, 12.99, 13.41]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Calculate correlation matrix
corr = df[['Open', 'High', 'Low', 'Close']].corr()

# Generate the heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix of Stock Prices')
plt.show()
```

```python
#Box Plot for All Data Columns
plt.figure(figsize=(12, 6))
df[['Open', 'High', 'Low', 'Close']].plot(kind='box', grid=True)
plt.title('Box Plot of Stock Prices')
plt.ylabel('Price')
plt.show()
```
